```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
'''
Your Code
'''
d = {} # create a dictionary

par = table.find_all('td')
length = len(par)
# print(length)
# print(par[0].get_attribute_list("style")[0] == 'white-space:nowrap;')
for i in range(len(par)):
    if (par[i].get_attribute_list("style")[0] == 'white-space:nowrap;'):
        particle = par[i].text
#         print(particle)
        d[particle] = {}
    else:
        if (par[i].get_attribute_list("valign")[0] == 'top'):
            continue
        if ("2019/" in par[i].text):
            month = par[i].text
#             print(month)
        else:
            if(par[i].text.strip() != ""):
                value = par[i].text
#                 print(value)
    d[particle][month] = value


print(d)
```
